```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from matplotlib.image import imread
import numpy as np
```

# Image Segmentation

```python
image=imread('Burung-Anis.png')
```

```python
image
```

```python
image.shape
```

```python
plt.imshow(image)
```

```python
X=image.reshape(-1,4)
X.shape
```

```python
X
```

```python
n_segments=2
kmeans=KMeans(n_clusters=n_segments,random_state=42).fit(X)
```

```python
kmeans.labels_
```

```python
kmeans.labels_.shape
```

```python
kmeans.cluster_centers_
```

```python
kmeans.cluster_centers_[kmeans.labels_]
```

```python
kmeans.cluster_centers_[kmeans.labels_].shape
```

```python
plt.imshow(kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape))
plt.axis('off')
```

```python
image1=imread('Kenari-Red-Factor.png')
plt.imshow(image1)
```

```python
image1.shape
```

```python
kmeans1=KMeans(n_clusters=2).fit(image1.reshape(-1,4))
plt.imshow(kmeans1.cluster_centers_[kmeans1.labels_].reshape(image1.shape))
```
